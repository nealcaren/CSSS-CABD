```python
from sklearn.datasets import load_digits
from sklearn.utils import shuffle

import pandas as pd
```

```python
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data)
df['y'] = digits.target
```

```python
df.head(10)
```

```python
df.describe()
```

```python
x_names = df.keys()[:-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchC

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
x_names = df.keys()[:-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[x_names], 
                                                    df['y'], 
                                                    test_size=0.2)
```

```python
l_model = LogisticRegression(fit_intercept=True)
l_model.fit(X_train,  y_train)
```

```python
confusion_matrix(y_train, l_model.predict(X_train))
```

```python
confusion_matrix(y_test, l_model.predict(X_test))
```

```python
knn_grid = {'n_neighbors': [3, 5, 10, 15]}
grid = GridSearchCV(KNeighborsClassifier(), knn_grid, cv=5)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
knn_model = grid.best_estimator_
```

```python
knn_model.fit(X_train, y_train)
```

```python
confusion_matrix(y_train, knn_model.predict(X_train))
```

```python
confusion_matrix(y_test, knn_model.predict(X_test))
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, knn_model.predict(X_test))
```

```python
accuracy_score(y_train, knn_model.predict(X_train))
```

```python
predicted_values = knn_model.predict(df[x_names])
```

```python
df['predicted_values'] = predicted_values
```
